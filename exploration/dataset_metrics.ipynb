```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
def discretize(df):
    for i in df.columns:
        if i != "Feature" and i != "Output":
            df[i] = pd.qcut(df[i], q=5,  labels=False, precision=0, duplicates='drop')
    return df
```

```python
df = pd.read_csv("full_dataset.csv")
```

```python
df = discretize(df)
```

```python
#Generate metrics for each feature
metrics = []
for i in df.columns:
    if i != "Feature" and i != "Output":
        metrics.append([i, df[i].min(), df[i].max(), df[i].mean(), df[i].std(), df[i].median(), df[i].mode()[0], df[i].quantile(0.25), df[i].quantile(0.75), df[i].skew(), df[i].kurtosis()])
metrics = pd.DataFrame(metrics, columns=["Feature", "Mean", "Std", "Median", "Mode", "Q1", "Q3", "Skew", "Kurtosis"])
```

```python
#Plot metrics
fig, ax = plt.subplots(1, 2, figsize=(20, 5))
sns.lineplot(x="Feature", y="Mean", data=metrics, ax=ax[0])
sns.lineplot(x="Feature", y="Std", data=metrics, ax=ax[1])
plt.show()
```

```python
new_metrics = metrics.loc[metrics['Mean'] < 1.8]

sns.lineplot(x="Feature", y="Mean", data=new_metrics)
plt.show()
```
